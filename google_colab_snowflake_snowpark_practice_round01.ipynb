```python
from snowflake.snowpark import Session
```

```python
import os
import json
```

```python
os.getcwd()
```

```python
with open('creds.json') as f:
  my_creds = json.load(f)
```

```python
my_sess = Session.builder.configs(my_creds).create()
```

```python
m_df = my_sess.sql("SHOW TABLES")
m_df.show()
```

```python
m_df_ue = my_sess.sql("SELECT * FROM EATS_ORDER_DETAILS_CSV_211832")
m_df_ue.show()
```

```python
l_df = m_df_ue.to_pandas()
l_df.info()
```
